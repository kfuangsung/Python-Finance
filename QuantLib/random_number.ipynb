```python
import QuantLib as ql 
import matplotlib.pyplot as plt 

plt.style.use('bmh')
```

```python
# random number generator

rng = ql.MersenneTwisterUniformRng()

xs = []
ys = []
for _ in range(2000):
    xs.append(rng.next().value())
    ys.append(rng.next().value())
```

```python
plt.scatter(xs, ys)
plt.tight_layout()
plt.show()
```

```python
# for random sequence
# be careful with dimensions

rng = ql.SobolRsg(1) # 1-dimension

xs = []
ys = []
for _ in range(2000):
    xs.append(rng.nextSequence().value()[0])
    ys.append(rng.nextSequence().value()[0])
```

```python
# obviously not random

plt.scatter(xs, ys)
plt.tight_layout()
plt.show()
```

```python
# need to use 2-dimension

rng = ql.SobolRsg(2) # 2-dimension

xs = []
ys = []
for _ in range(2000):
    x, y = rng.nextSequence().value()
    xs.append(x)
    ys.append(y)
```

```python
plt.scatter(xs, ys)
plt.tight_layout()
plt.show()
```
