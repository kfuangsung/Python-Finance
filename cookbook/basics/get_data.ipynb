## Intrinio

```python
intrinio_sdk.ApiClient().configuration.api_key['api_key'] = apikey.intrinio_api_key
security_api = intrinio_sdk.SecurityApi()
```

```python
r = security_api.get_security_stock_prices(identifier='AAPL')
```

```python
data = [x.to_dict() for x in r.stock_prices]
data = pd.DataFrame(data).sort_values('date')
data.set_index('date', inplace=True)
```
