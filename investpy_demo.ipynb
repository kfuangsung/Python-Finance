# Search Quotes

```python
search_res = investpy.search_quotes(text='apple', 
                                    products=['stocks'],
                                    countries=['United States'],
                                    n_results=5)
for res in search_res:
    print(res)
    print('-'*50)
```

```python
search_res[0].symbol
```

```python
search_res[0].retrieve_currency()
```

```python
search_res[0].retrieve_information()
```

```python
search_res[0].retrieve_technical_indicators(interval='daily')
```
